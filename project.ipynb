## O Programa Universidade para Todos (Prouni)

O Programa Universidade para Todos (Prouni) é uma iniciativa do governo brasileiro que visa ampliar o acesso ao ensino superior por meio da concessão de bolsas de estudo integrais e parciais em instituições privadas. Criado em 2004, o Prouni beneficia estudantes de baixa renda, possibilitando que eles ingressem em cursos de graduação sem arcar com os custos das mensalidades.

## A análise do Prouni

A análise do Prouni é fundamental para compreender seu impacto na educação superior. Uma das principais conclusões obtidas interpretando os dados é a presença majoritária do sexo feminino sendo beneficiaria do programa. Em resumo, o Prouni é uma política educacional relevante para a inclusão social e o acesso à educação superior no Brasil. Sua avaliação contínua é essencial para aprimorar seus resultados e garantir que mais estudantes tenham oportunidades de formação acadêmica.

##### Nos dados originais, não é considerada a idade. Logo, vamos desconsiderar a coluna "idade" e mais adiante vamos fazer cálculo da mesma, ao momento da obtenção da bolsa

```python
df = dados[['ANO_CONCESSAO_BOLSA', 'NOME_IES_BOLSA',
       'TIPO_BOLSA', 'MODALIDADE_ENSINO_BOLSA', 'NOME_CURSO_BOLSA',
       'NOME_TURNO_CURSO_BOLSA',
       'SEXO_BENEFICIARIO_BOLSA', 'RACA_BENEFICIARIO_BOLSA', 'DT_NASCIMENTO_BENEFICIARIO',
        'BENEFICIARIO_DEFICIENTE_FISICO',
       'REGIAO_BENEFICIARIO_BOLSA', 'SIGLA_UF_BENEFICIARIO_BOLSA']].copy()
```

```python
df.isnull().sum()
```

##### As colunas com dados faltantes vão ser consideradas para algumas analises. Note que a porcemtagem dos dados faltantes é menos do que o 1% dos dados. Logo, vamos a deconsiderar tais dados. 

```python
df = df.dropna(subset=df.columns)
df.shape
```

```python
df.head()
```

##### Vamos considerar a idade do bolsista como os anos cumpridos no ano da obtenção da bolsa

```python
df["Ano_nascimento"] = dados["DT_NASCIMENTO_BENEFICIARIO"].str.extract('(\\d+)').astype(int)
df["IDADE"] = df['ANO_CONCESSAO_BOLSA'] - df['Ano_nascimento']

df = df.drop("Ano_nascimento", axis=1)
```

```python
df["IDADE"].describe()
```

##### Note-se que temos alguns dados de idade discordantes com o ambiente de estudo. Consideramos que tais dados  são  consequências de erros de digitação. Assim, vamos proceder a uma limpeza dos dados, a partir da idade, e iremos considerar só dados com idades entre os 15 e os 70 anos.

```python
mask_min = df['IDADE']>14
mask_max = df['IDADE']<71
df = df[mask_min & mask_max]
```

##### Um outro tramtamento nos dados que iremos fazer é referente ao nome de alguns cursos, que apresentam algumas variações em anos específicos.

```python
mapeo_correccion = {'AdministraAAo': 'Administracao'}
mapeo_correccion2 = {'CiAancias ContAbeis': 'Ciencias Contabeis'}
# Corregir los valores incorrectos utilizando el método replace()
df['NOME_CURSO_BOLSA'] = df['NOME_CURSO_BOLSA'].replace(mapeo_correccion)
df['NOME_CURSO_BOLSA'] = df['NOME_CURSO_BOLSA'].replace(mapeo_correccion2)
```

##### Iniciamos considerando alguns gráficos para obter uma ideia sobre a distribuição das bolsas em cada um dos campos de interese.

```python
cols = ['REGIAO_BENEFICIARIO_BOLSA', 'TIPO_BOLSA', 'MODALIDADE_ENSINO_BOLSA', 'SEXO_BENEFICIARIO_BOLSA']

cols_ = ['região do beneficiario', 'tipo de bolsa', 'modalidade', 'sexo do beneficiario']
fig, axs = plt.subplots(2, 2, figsize=(15, 18))


for i in range(len(cols)):
    try:
        
        row = i // 2
        col = i % 2

        
        caso = df[cols[i]].value_counts()

        
        caso.plot.pie(autopct='%1.1f%%', startangle=90, ax=axs[row, col])
        axs[row, col].set_title(f'Distribucao das bolsas por {cols_[i]}')
        axs[row, col].set_ylabel('')
        plt.gca().set_aspect('equal')

    except:
        continue


plt.tight_layout()


plt.show()
```

```python
total = len(df)
porcentajes = df['RACA_BENEFICIARIO_BOLSA'].value_counts() / total * 100

# Crear el gráfico de barras
plt.figure(figsize=(9, 6))
sns.barplot(x=porcentajes.index, y=porcentajes.values, palette='viridis')

# Establecer etiquetas y título
plt.xlabel('Raça')
plt.title('Porcentagem dos Beneficiarios por Raça')

for index, value in enumerate(porcentajes):
    plt.text(index, value + 0.5, f'{value:.2f}%', ha='center', va='bottom', fontsize=10)

plt.tick_params(axis= 'both', bottom=True, labelbottom=True, left = False, labelleft= False)
# Mostrar el gráfico
plt.show()
```

O gráfico mostra a distribuição de bolsas de estudo entre pessoas de diferentes cores ou raças no Brasil, segundo os dados estudados. Esses dados evidenciam os efeitos da política de inclusão e de cotas nas instituições parceiras do Prouni, que visa garantir o acesso de grupos historicamente excluídos ou discriminados à educação. No entanto, ainda há desafios a serem superados, como a permanência e a qualidade da formação desses estudantes.

### Programas prediletos por região

```python
regioes = df['REGIAO_BENEFICIARIO_BOLSA'].unique()

regioes = regioes.astype(object)


regioes = regioes[~pd.isnull(regioes)]
```

```python

num_filas = (len(regioes) + 1) // 2


fig, axs = plt.subplots(num_filas, 2, figsize=(15, 6 * num_filas))


for i, regiao in enumerate(regioes):
    try:
        
        row = i // 2
        col = i % 2

        # Dados por região
        dados_regiao = df[df['REGIAO_BENEFICIARIO_BOLSA'] == regiao]
        top_10_cursos = dados_regiao['NOME_CURSO_BOLSA'].value_counts().head(10)

        
        axs[row, col].bar(top_10_cursos.index, top_10_cursos.values)
        axs[row, col].set_title(f'Top 10 Cursos no {regiao}')
        axs[row, col].set_xlabel('Curso')
        axs[row, col].set_ylabel('Número de Bolsas')
        axs[row, col].tick_params(axis='x', rotation=75)

    except:
        continue

# Dados do pais
top_10_cursos = df['NOME_CURSO_BOLSA'].value_counts().head(10)


axs[2, 1].bar(top_10_cursos.index, top_10_cursos.values)
axs[2, 1].set_title(f'Top 10 Cursos no Pais')
axs[2, 1].set_xlabel('Curso')
axs[2, 1].set_ylabel('Número de Bolsas')
axs[2, 1].tick_params(axis='x', rotation=75)


plt.tight_layout()


plt.show()
```

```python
top5_cursos= df['NOME_CURSO_BOLSA'].value_counts().head(5).index.tolist()
df_top5_cursos = df[df['NOME_CURSO_BOLSA'].isin(top5_cursos)]

# agrupar os dados
df_group_top5cursos_anos = df_top5_cursos.groupby(["ANO_CONCESSAO_BOLSA", 'NOME_CURSO_BOLSA']).size().unstack(fill_value=0)

# graficar os dados
plt.figure(figsize=(12, 6))
sns.lineplot(data=df_group_top5cursos_anos, markers=True, dashes=False)
plt.xlabel('Ano')
plt.ylabel('Número de Bolsas')
plt.title('Cursos ----- ao longo dos anos')
plt.legend(title='Curso', loc='upper left')
plt.show()
```

Este é um excelente gráfico que mostra o número de bolsas oferecidas por cursos ao decorrer dos anos da database. Observa-se o aumento geral no número de bolsas oferecidas e também a cada curso.

```python
# Agrupa os dados por ano e nome do curso, e conta as ocorrências
bolsas_por_ano = df.groupby(['ANO_CONCESSAO_BOLSA', 'NOME_CURSO_BOLSA']).size()

# Para cada ano, seleciona os top 3 cursos com mais bolsas
top_3_bolsas_por_ano = bolsas_por_ano.groupby(level=0).nlargest(3).reset_index(level=0, drop=True)

# Exibe os resultados
print(top_3_bolsas_por_ano)
# Converte a série em um DataFrame
top_3_bolsas_df = top_3_bolsas_por_ano.reset_index()
top_3_bolsas_df.columns = ['ANO_CONCESSAO_BOLSA', 'NOME_CURSO_BOLSA', 'COUNT']

# Reformata os dados para exibição em uma tabela
pivot_df = top_3_bolsas_df.pivot(index='ANO_CONCESSAO_BOLSA', columns='NOME_CURSO_BOLSA', values='COUNT')

# Preenche os valores NaN com 0
pivot_df = pivot_df.fillna(0)

# Exibe a tabela
print(pivot_df)
```

### Programas prediletos por Faixa Etária

```python

palettes = ['coolwarm', 'Spectral', 'viridis', 'cubehelix', 'rocket']


grupos_anos = ['0-19','20-24','25-29','30-34', '35-39','40-44',
          '45-49','50-54', '55-59', '60-64', '65-70']
num_filas = (len(grupos_anos) + 1) // 2


fig, axs = plt.subplots(num_filas, 2, figsize=(15, 6 * num_filas))


for i, grupo in enumerate(grupos_anos):
    try:
        #  coordenadas do subplot
        row = i // 2
        col = i % 2

        # datos por região
        dados_grupo = df[df['GRUPO_IDADE'] == grupo]
        top_10_cursos = dados_grupo['NOME_CURSO_BOLSA'].value_counts().head(10)

        #  gráfico de barras do subplot correspondiente
        axs[row, col].bar(top_10_cursos.index, top_10_cursos.values, color=sns.color_palette(palettes[i % len(palettes)], len(top_10_cursos)))
        axs[row, col].set_title(f'Top 10 curso em {grupo} anos')
        axs[row, col].set_xlabel('Cursos prediletos')
        axs[row, col].set_ylabel('Número de Bolsas')
        axs[row, col].tick_params(axis='x', rotation=75)

    except:
        continue

# Ajustar subplots
plt.tight_layout()


plt.show()
```

Podemos perceber alguns pontos interessantes através dos gráficos: 
  A repetição entre os 3 cursos mais escolhidos, sendo eles Pedagogia, Direito e Administração; tendo como exceção, o primeiro grupo de faixa etária, que surge Enfermagem, 
  e com o mais procurado em geral sendo Pedagogia.
