```python
n_dados = dados[['ANO_CONCESSAO_BOLSA', 'NOME_IES_BOLSA',
       'TIPO_BOLSA', 'MODALIDADE_ENSINO_BOLSA', 'NOME_CURSO_BOLSA',
       'NOME_TURNO_CURSO_BOLSA',
       'SEXO_BENEFICIARIO_BOLSA', 'RACA_BENEFICIARIO_BOLSA',
        'BENEFICIARIO_DEFICIENTE_FISICO',
       'REGIAO_BENEFICIARIO_BOLSA', 'SIGLA_UF_BENEFICIARIO_BOLSA', 'idade']].copy()
```

```python
n_dados.columns = ['ANO', 'NOME_IES',
       'TIPO_BOLSA', 'MODALIDADE', 'NOME_CURSO', 'TURNO',
       'SEXO_BENEFICIARIO', 'RACA_BENEFICIARIO',
        'BENEFICIARIO_DEFICIENTE_FISICO',
       'REGIAO_BENEFICIARIO', 'SIGLA_UF_BENEFICIARIO', 'idade']
```

```python
n_dados["idade"] = n_dados["idade"].astype(int)
```

```python
n_dados['REGIAO_BENEFICIARIO'].value_counts()
```

```python
cols = ['REGIAO_BENEFICIARIO', 'TIPO_BOLSA', 'MODALIDADE', 'SEXO_BENEFICIARIO',
        'RACA_BENEFICIARIO', 'BENEFICIARIO_DEFICIENTE_FISICO']


for i in range(len(cols)):
    caso = n_dados[cols[i]].value_counts()
    caso.plot.pie(autopct='%1.1f%%', startangle=90)
    plt.axis('equal')  # Hace que el diagrama de torta sea circular
    plt.title(f'Distribucao das bolsas por {cols[i]}')
    plt.ylabel('')  # Eliminar la etiqueta del eje y
    plt.show()
    
# fig.text(0.13, 0.95, 'Prouni - Overall',
         #fontsize=18, fontweight='bold', fontfamily='garamond') 
# plt.show()
```
